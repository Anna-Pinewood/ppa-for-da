###Введение

Текст статьи:
https://link.springer.com/content/pdf/10.1007/s00779-018-1121-x.pdf

Введение: Основная цель статьи - исследование корреляции между настроениями инвесторов и ценой акций. Анализ настроений ивесторов основан на данных из китайской платформы Guba, на которой инвесторы делятся новостями и мнениями об акциях.

Рассмотрим далее этапы анализа данных, которые можно выделить в этой статье, в соответствии с методологией CRISP-DM.

###Понимание бизнеса (Business Understanding)

Три первых раздела статьи - "Background", "Literature analysis", "Guba sentiment hypotheses" можно отнести к этому этапу, в нём авторы говорят об исследованиях влияния эмоций инвесторов на рынок акций, и на основании этого формируют гипотезы. 

Также можно трактовать как оценку ситуации:
*  анализ исследований проведённых ранее
*  утверждение о развитие технологий
*  анализ структуры рынка (80% инвесторов в Китае, остальные в США)

Декларирование цели анализа данных:

*In this work, we try to investigate the correlations between the Guba-based sentiment and the stock market trends in China. We propose a novel method to measure investors’ online sentiment. According to the findings, we finally build a stock recommendation system based on Guba-sentiment.*




###Понимание данных (Data Understanding)
Раздел "4.1 Datasets" посвящен сбору исходных данных. В статье сказано, чтобы был создан парсер, собравший большое количество данных с форумов Guba, а также авторы собрали релевантные новости, данные рынка и информацию о компаниях.

Финансовые данные по акция (цена открытия/закрытия и т.д.) были получены из базы данных RESSET.

###Подготовка данных (Data Preparation)
В разделе "4.2 Guba investor sentiment" рассматривается получение итогового набора данных. Необходимо было создать словарь специальных слов, с помощью которых инвесторы выражают своё настроение, относительно происходящего на рынке.

В конечно счёте был составлен словарь GubaLex, включающий 47274 основных слова и 1429 биржевых термина. Он также всключает в себя ещё два словаря  GL-Bull и GL-Bear, содержащих 6603 и 11338 слов соответственно. Был создан модуль для распознавания и добавления новых слов.

Нерелевантные записи с форумов были отфильтрованы, метод фильтрации не уточняется. В разделе "4.2.3 Sentiment score computation" описана методика рассчета итоговой метрики, показывающей настроение каждого поста.

###Моделирование (Modeling)
Раздел "4.3 Results" можно назвать переходным между этапом моделирования и этапом подготовки данных, в нём, основываясь на подготовленных ранее данных, проверяются базовые гипотезы, на основе которых принимается решение о рентабельности создания модели машинного обучения.

Раздел 5.1 посвящен созданию модели машинного обучения и созданию рекомендательной системы. В качестве алгоритма машинного обучения была выбрана Вероятностная неройнная сеть (Probabilistic neural network, PNN).

Авторы считают модель эффективной, т.к. она показала точность 86.3% на тестовой выборке.

Информация относительно следующих шагов не представлена.

###Оценка (Evaluation)

В разделе 5.2, с одноимённым названием, оцениваются результаты работы модели. Согласно стратегии система продавала и покупала акции, что в конечном счете позволило ей увеличить капитал в 8 раз.
