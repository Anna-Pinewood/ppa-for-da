# Обзор статьи "Which CNNs and Training Settings to Choose for Action Unit  Detection? A Study Based on a Large-Scale Dataset"



Статья: https://arxiv.org/pdf/2111.08320.pdf

# Понимание бизнеса

Во введении говорится о превосходстве CNN в распознавании мимики человека над признаками созданными вручную (Handcrafted features).

Целью статьи является сравнение влияния различных CNN архитектур и настроек обучения на производительность AU (Блок распознавания действий).




# Понимание данных

Для получения надежных и поддающихся обощению результатов, используется крупномасштабный набор данных состоящий из 55 тысяч видеороликов (из более чем 90 стран). 

В видероликах участники просматривают коммерческую рекламу. Видео участников были сняты в различных условиях записи (например, в дикой природе) и с полностью спонтанными выражениями лиц. Это отличается от других существующих наборов данных с относительно ограниченными предметами, условиями записи и скудным демографическим разнообразием.

Видео были аннотированы для различных AU группой обученных кодировщиков FACS (Системы кодирования действий лица).

# Подготовка данных

Для экспериментов набор данных делится на 40,9 тысяч видео для обучения, 5,9 тысяч для проверки и 8,2 тысяч для тестирования.

Собранные видео были аннотированы на наличие AU с использованием обученных кодеров FACS. Кроме того, видео были помечены по полу (55% женщин, 37% мужчин, 8% неопределенных) и этнической принадлежности (37% европеоидов, 24% восточноазиатских, 14% южноазиатских, 13% латинских, 9% африканских, 3% других).



# Моделирование

Во всех экспериментах CNN инициализируется случайным образом из одного и того же начального числа и обучается в течение 600 эпох с использованием оптимизатора Adagrad с начальной скоростью обучения, установленной на 0,005. 

Для каждой эпохи 300 пакетов отбираются для обучения и 300 пакетов для проверки. Обучающие пакеты дополняются случайным переворачиванием, сдвигом, масштабированием и т. Д. 

Для тестирования лучшая валидационная модель проверяется на 3000 пакетах, отобранных из набора для тестирования. Точность используется для оценки производительности, поскольку есть равное количество положительных и отрицательных примеров в партиях проверки и тестирования.
