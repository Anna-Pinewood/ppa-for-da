Для предобработки табличных данных воспользуемся библиотекой panddas

Требуется импортировать библиотеку

```python
import pandas as pd
```

В нашем уроке воспользуемся тестовыми данными 

```python
df = pd.DataFrame([[18, 160, "1"], [19, 170, "2"], [20, 175, None], [21, 170, "4"], [22, 160, "5x"]])
df
```

Переименуем колонки, чтобы в дальнейшем с ними было удобнее работать

```python
col_name = ['age', "height", "num"]
df.set_axis(col_name, axis = 'columns', inplace = True)
```

Для удаления строк у которых отсутствует хотя бы одно значение воспользуемся методом dropna. Стоит обратить внимание что вызов данного метода никак не изменяет на df поэтому здесь мы увидим результат, но не сохраним его

```python
df.dropna()
```

Если требуется удалить строки в которых отсутствуют конкретные значения, можно воспользоваться аргументом subset

```python
df.dropna(subset=['height'], inplace=True)
df
```

Рассмотрим тип колонки num. Как видно ее тип object, так как существуют строки которые невозможно привести к числовому типу

```python
df['num']
```

Чтобы поменять тип воспользуемся методом astype

```python
df['num'].astype('int64')
```

Получим ошибку. Для того чтобы исправить ее, требуется удалить None и не числовые строки из выборки.

```python
df = df[df['num'].isnull() != True]
df
```

```python
df = df[pd.to_numeric(df['num'], errors='coerce').notnull()]
df
```

```python
df = df.astype({'num': 'int32'})
df
```
