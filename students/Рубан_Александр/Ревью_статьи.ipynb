##Цель
Обучение и примение алгоритмов обработки естественного языка (NLP) для автоматической радиолого-артроскопической корреляции менискальных разрывов из отчетов об обследованиях.

Рассмотрим далее этапы анализа данных, которые можно выделить в этой статье, в соответствии с методологией CRISP-DM.

##Понимание бизнеса (Business Understanding)
Оценка диагностических показателей рентгенологов важна для демонстрации клинической ценности и определения областей для улучшения, но ручной ретроспективный обзор отчетов по радиологии и артроскопии занимает много времени и требует знаний по конкретным областям.

##Понимание данных (Data Understanding)

Набор данных представлет собой отчеты МРТ коленного сустава на протяжении года.  Собранные данные включали все МРТ коленного сустава, полученные за годичный период с 1 января 2018 г. по 31 декабря 2018 г., в общей сложности 3631 отчет МРТ коленного сустава от 3448 уникальных пациентов. Правильная идентификация этих результатов позволяет автоматически выявлять разногласия между отчетами. 

Отчеты МРТ имеют полуструктурированные заголовки и менялись в зависимости от использемого шаблона, но текстовое содержимое в каждом заголовке не было структурировано.

##Подготовка данных (Data Preparation)

Полные отчеты, включая результаты исследования и слепок с МРТ коленного сустава, были предварительно обработаны для ввода в контролируемые модели машинного обучения. 

Во-первых, предложения, содержащие «мениск» или «рог», были извлечены из текста с использованием библиотеки операций с регулярными выражениями re в Python. 

Отчеты, в которых эти термины не упоминались, были исключены из дальнейшего анализа, поскольку такие отчеты относились к случаям, когда в прошлом была проведена полная артропластика, резекция опухоли или целенаправленные исследования нервных структур. 

Это исключило 38 отчетов МРТ от 31 пациента (1% отчетов), что дало окончательный набор данных из 3593 отчетов МРТ от 3417 пациентов.

Выборка была разделена в соотошении 80/20.

Разделение произошло на уровне пациента без пересечения пациентов между тренировочной и тестовой наборами. Для 108 артроскопий, выполненных для МРТ в этом наборе тестов, отчеты артроскопии были вручную аннотированы для медиального и латерального разрыва мениска.

В наборе данных МРТ средний возраст пациентов составлял 47 лет (стандартное отклонение ± 18 лет). 
- 254 (7%) МРТ были получены от педиатрических пациентов (<18 лет); 
- 1897 (53%) МРТ были получены от пациентов, идентифицированных как женщины;
- 1696 (47%) были получены от пациентов, идентифицировавших себя как мужчин.

Распространенность медиальных и боковых разрывов мениска в полном наборе данных МРТ коленного сустава составила 44% и 22% соответственно.

##Моделирование (Modeling)

Предварительно обработанные отчеты из обучающего набора затем передавались в качестве входных данных для обучения логистической регрессии, SVM и случайного леса для классификации разрывов мениска с использованием библиотеки sklearn ​​в Python.

##Оценка (Evaluation)



Модели логистической регрессии, SVM и случайного леса, оцененные для обнаружения разрывов мениска в отчетах МРТ коленного сустава с произвольным текстом, показали стратифицированную пятикратную кросс-валидацию в диапазоне от 0,94 до 0,95 и оценку F1 в диапазоне от 0,86 до 0,94
